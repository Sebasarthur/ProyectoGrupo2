```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/content/global_unemployment_data.csv')

data = pd.DataFrame(df)


```

```python
#data frame para filtrar por los paises de centro america

dataCentroAmerica = data.loc[(data['country_name']=='Belize') | (data['country_name']=='Costa Rica')|(data['country_name']=='El Salvador')|(data['country_name']=='Guatemala')|(data['country_name']=='Honduras')|(data['country_name']=='Nicaragua')|(data['country_name']=='Panama')]
dataCentroAmerica.head()
```

```python
#data frame para filtrar por los paises de norte america

dataNorteAmerica = data.loc[(data['country_name']=='Canada') | (data['country_name']=='Mexico')|(data['country_name']=='United States')]
dataNorteAmerica.head()
```

```python
# Lista de países de Sudamérica
paises_suramerica = [
    'Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador',
    'Guyana', 'Paraguay', 'Peru', 'Suriname', 'Uruguay', 'Venezuela','Trinidad and Tobago'
]

#dataframe paises suramerica
dataSurAmerica = data.loc[data['country_name'].isin(paises_suramerica)]

dataSurAmerica.head()
```

```python
#Pais de centroamerica con mayor desempleo en el 2024

idx_max = dataCentroAmerica['2024'].idxmax()

# Usa el índice para obtener la fila completa
dataMaxCen = dataCentroAmerica.loc[idx_max]

dataMaxCen
```

```python
#Pais de norte con mayor desempleo en el 2024

idx_max = dataNorteAmerica['2024'].idxmax()

# Usa el índice para obtener la fila completa
dataMaxNort = dataNorteAmerica.loc[idx_max]

dataMaxNort
```

```python
#Pais de sur con mayor desempleo en el 2024

idx_max = dataSurAmerica['2024'].idxmax()

# Usa el índice para obtener la fila completa
dataMaxSur = dataSurAmerica.loc[idx_max]

dataMaxSur
```
